## Unpivote this data from wide to long format

```python
# Unpivoting dataframe
data_1 = pd.melt(data,id_vars = sector, value_vars = year, var_name='year', value_name = 'FDI in US$ Millions')
data_1
```

```python
# Sorting the sectors
sorted = data_1.sort_values(['Sector','year'])
sorted
```

```python
# Check the description of sectors
pd.set_option('display.max_rows', None)
sorted.groupby('Sector').describe().round(2)
```

```python
# Rename sectors
sorted = sorted.replace(["CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects"
                              ,"SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)"
                              ,'TEA AND COFFEE (PROCESSING & WAREHOUSING COFFEE & RUBBER)']
                             ,["CONSTRUCTION DEVELOPMENT","SERVICES SECTOR",'TEA AND COFFEE'])
```

```python
# Lets check the sector wise total FDI from 2000-17
total_fdi_sector = sorted.groupby('Sector').sum().sort_values(by = 'FDI in US$ Millions',ascending = False)
total_fdi_sector['% of total FDI'] = round(total_fdi_sector*100/total_fdi_sector.sum(),2)
total_fdi_sector
```

```python
# Visualisation of sectorwise total FDI
plt.figure(figsize=(8,30))
sns.barplot(x='FDI in US$ Millions',y = total_fdi_sector.index ,data = total_fdi_sector)
plt.xlabel('FDI in US$ Millions',fontsize=13)
plt.ylabel('Sectors',fontsize=13)
plt.title('Sectorwise total FDI',fontsize=16)
plt.grid()
plt.show()
```

## Check Best and worst performing sectors

```python
# Lets check Top 10 performing secotrs
top_10 = total_fdi_sector.head(10)
top_10
```

```python
# Lets visualise Top 10 performing secotrs
plt.figure(figsize=(15,8))
plots = sns.barplot(x = top_10.index, y='FDI in US$ Millions' ,data = top_10)
for bar in plots.patches:
  plots.annotate(format(bar.get_height(),'.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.xlabel('Sectors',fontsize=13)
plt.ylabel('FDI in US$ Millions',fontsize=13)
plt.xticks(rotation = 90)
plt.text(7,55000, "(Amounts in US$ Millions)", fontsize = 15)
plt.title('Top 10 performing secotrs',fontsize=16)
plt.show()
```

```python
#pie chart to visualise percentage share of FDI among top 10 sectors
plt.figure(figsize=(20,8))
colors = sns.color_palette('pastel')[0:7]
textprops = {'fontsize':13}
plt.pie(top_10['FDI in US$ Millions'],labels=top_10.index,colors=colors, autopct="%1.1f%%",textprops = textprops)
plt.axis('equal')
plt.title('SHARE AMONG TOP 10 SECTORS', fontsize = 20)
plt.show()
```

```python
# Lets check worst 10 performing secotrs
least_10 = total_fdi_sector.sort_values(by='FDI in US$ Millions', ascending=True).head(10)
least_10
```

```python
# Lets visualise worst 10 performing secotrs
plt.figure(figsize=(15,8))
plots = sns.barplot(x = least_10.index, y='FDI in US$ Millions' ,data = least_10)
for bar in plots.patches:
  plots.annotate(format(bar.get_height(),'.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.xlabel('Sectors',fontsize=13)
plt.ylabel('FDI in US$ Millions',fontsize=13)
plt.xticks(rotation = 90)
plt.text(0,150, "(Amounts in US$ Millions)", fontsize = 15)
plt.title('Worst 10 secotrs',fontsize=16)
plt.show()
```

```python
#pie chart to visualise percentage share of FDI among worst 10 sectors
plt.figure(figsize=(20,8))
colors = sns.color_palette('pastel')[0:7]
textprops = {'fontsize':13}
plt.pie(least_10['FDI in US$ Millions'],labels=least_10.index,colors=colors, autopct="%1.1f%%",textprops = textprops)
plt.axis('equal')
plt.title('SHARE AMONG WORST 10 SECTORS', fontsize = 20)
plt.show()
```

# FDI Inflow Yearwise

```python
year_inflow = data_1[['year', 'FDI in US$ Millions']]
year_inflow = year_inflow.groupby('year').sum()
year_inflow['% growth over previous year'] = round(year_inflow.pct_change()*100,2)
year_inflow
```

```python
#plotting to show Year by Year FDI Inflow
year_inflow.plot.line(y='FDI in US$ Millions',figsize = (20,9))
plt.xlabel('Year', fontsize = 13)
plt.ylabel('FDI in US$ Millions', fontsize = 13)
plt.title('FDI Inflow', fontsize = 15)
plt.show()
```

## From above graph we can say that
1. from 2000-01 upto 2005-06 there is not much inflow in India.
2. But After 2005-06 there is hike in inflow 
3. we face some downfall but overall there is growth in FDI inflows.
