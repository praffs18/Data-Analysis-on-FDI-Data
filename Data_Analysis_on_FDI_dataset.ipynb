# FDI Analysis

```python
# Import all necessary libraries
import pandas as pd
import numpy as np

# Import Visualisation libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Import Warnings
import warnings
warnings.filterwarnings('ignore')
```

# Data loadning & Inspection

```python
# Load the data
url = 'https://raw.githubusercontent.com/praffs18/Data-Analysis-on-FDI-Data/main/FDI_in_India.csv'
data = pd.read_csv(url)
```

```python
# lets check what are the columns we have
data.columns
```

So, in this data set their are two types of column
1. Sector & Yearwise investment

```python
sector = ['Sector']
year = ['2000-01', '2001-02', '2002-03', '2003-04', '2004-05',
       '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11',
       '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
```

```python
# lets check information of the data
data.info()
```

```python
# Lets check the description of the data
data.describe().round(2)
```

```python
# Lets check is there in missing value in data
data.isnull().sum()
```

```python
# Lets check any duplicate values in data
data.duplicated().sum()
```

## From this data inspection we can say that 
1. This data set is not having any missing values as well as duplicate values in it
2. All the numerial values in this dataset are in US$ Millions
3. Also we can say that this data is of yearwise investment in all sectors

## Unpivote this data from wide to long format

```python
# Unpivoting dataframe
data_1 = pd.melt(data,id_vars = sector, value_vars = year, var_name='year', value_name = 'FDI in US$ Millions')
data_1
```

```python
# Sorting the sectors
sorted = data_1.sort_values(['Sector','year'])
sorted
```

```python
# Check the description of sectors
pd.set_option('display.max_rows', None)
sorted.groupby('Sector').describe().round(2)
```

```python
# Lets check the sector wise total FDI from 2000-17
total_fdi_sector = sorted.groupby('Sector').sum().sort_values(by = 'FDI in US$ Millions',ascending = False)
total_fdi_sector
```

```python
# Visualisation of sectorwise total FDI
plt.figure(figsize=(8,30))
sns.barplot(x='FDI in US$ Millions',y = total_fdi_sector.index ,data = total_fdi_sector)
plt.xlabel('FDI in US$ Millions',fontsize=13)
plt.ylabel('Sectors',fontsize=13)
plt.title('Sectorwise total FDI',fontsize=16)
plt.grid()
plt.show()
```
